```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split as tts

a=pd.read_csv(r"C:\users\jaysu\Desktop\DS&ML\priority_1\ram_prices.csv")
data=pd.DataFrame(a).values
X=data[:,0].reshape(-1,1)
y=data[:,1].reshape(-1,1)

plt.plot(X,y)
plt.title('raw data')
plt.figure()
plt.hist(y, bins=100)
plt.title('histogram of raw data')
plt.figure()
plt.plot(X,np.log(y))
plt.title('log')
plt.figure()
plt.hist(np.log(y), bins=100)
plt.title('histogram after log')

X_train,X_test,y_train,y_test=tts(X,np.log(y),random_state=2)
lr=LinearRegression()
lr.fit(X_train,y_train)
print('test accuracy after log:',lr.score(X_test,y_test))
```
